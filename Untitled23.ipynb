```python
df = df.drop_duplicates()
print(df)
```

```python
# Total units sold by product
total_units_sold = df.groupby('Product Name')['Units Sold'].sum().sort_values(ascending=False)
highest_units_sold_product = total_units_sold.idxmax()
highest_units_sold_value = total_units_sold.max()
print(f"The product with the highest total units sold is {highest_units_sold_product} with {highest_units_sold_value} units sold.")
```

```python
# Total units sold by company
units_sold = df.groupby('Company')['Units Sold'].sum()
# Find the company with the highest export units
top_company = units_sold.idxmax()
top_company_units = units_sold.max()
total_units = units_sold.sum()
company_percentage = (units_sold / total_units) * 100
print(f"The company with the most units exported is {top_company} with {top_company_units} units sold.")
print(f"Percentage contribution of each company: {company_percentage}")
```

```python
# Units received by destination port
destination_port_units = df.groupby('Destination Port')['Units Sold'].sum().sort_values(ascending=False)
top_destination_port = destination_port_units.idxmax()
top_destination_units = destination_port_units.max()
print(f"The destination port with the highest number of units sold is {top_destination_port} with {top_destination_units} units.")
```

```python
# Average unit price per product
average_unit_price = df.groupby('Product Name')['unit_price'].mean().sort_values(ascending=False)
print(f"Average unit price per product: {average_unit_price}")
```

```python
# Total export value by product and company
df['Export Value'] = df['Units Sold'] * df['unit_price']
highest_export_value = df.groupby('Product Name')['Export Value'].sum().sort_values(ascending=False)
top_export_product = highest_export_value.idxmax()
top_export_value = highest_export_value.max()
print(f"The product with the highest export value is {top_export_product} with a value of {top_export_value}.")
```

```python
# Profit' is provided or can be calculated
total_profit = df.groupby('Product Name')['Profit per unit'].sum().sort_values(ascending=False)
most_profitable_product = total_profit.idxmax()
most_profitable_value = total_profit.max()
average_profit_per_unit = df.groupby('Product Name')['Profit per unit'].mean().sort_values(ascending=False)
print(f"The product generating the most total profit is {most_profitable_product} with a profit of {most_profitable_value}.")
print(f"Average profit per unit for each product: {average_profit_per_unit}")
```

```python
#  total export value by country
export_by_country = df.groupby('Export Country')['Export Value'].sum().sort_values(ascending=False)
top_export_country = export_by_country.idxmax()
top_export_country_value = export_by_country.max()
print(f"The country contributing the most to the total export value is {top_export_country} with a value of {top_export_country_value}.")
```

```python
# The year with the highest export
df['Year'] = pd.DatetimeIndex(df['Date']).year
yearly_exports = df.groupby('Year')['Units Sold'].sum().sort_values(ascending=False)
highest_export_year = yearly_exports.idxmax()
highest_export_units = yearly_exports.max()
print(f"The year with the highest export activity is {highest_export_year} with {highest_export_units} units sold.")
```

```python
# Line chart for export values over time
df['Date'] = pd.to_datetime(df['Date'])
monthly_export_values = df.groupby(df['Date'].dt.to_period('M'))['Export Value'].sum()
monthly_export_values.plot(kind='line', title='Monthly Export Values Over Time')
plt.xlabel('Month')
plt.ylabel('Export Value')
plt.show()
```

```python
# Bar chart for top 5 companies by export value
top_companies = df.groupby('Company')['Export Value'].sum().sort_values(ascending=False).head(5)
top_companies.plot(kind='bar', title='Top 5 Companies by Export Value')
plt.xlabel('Company')
plt.ylabel('Export Value')
plt.show()
```

```python
# Stacked bar chart for products exported to destination ports
product_destination = df.groupby(['Destination Port', 'Product Name'])['Units Sold'].sum().unstack()
product_destination.plot(kind='bar', stacked=True, title='Products Exported to Destination Ports')
plt.xlabel('Destination Port')
plt.ylabel('Units Sold')
plt.show()
```
